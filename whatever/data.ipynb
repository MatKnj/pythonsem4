## dygresja - wyrazenia lambda

```python
lst = [3, 2, 2, 1, 5, 3, 5, 6, 9, 0]
```

```python
#   dygresja^2:     da sie tez nadpisywać klasy wbudowane w pythona
```

```python
sorted(lst)
```

```python
lst1 = [('Adam', 25), ('Eve', 25), ('Charlie', 18), ('juz nie wiem kto', 99)]
```

```python
sorted(lst1)
```

```python
def get_age(v):
    return v[1]
```

```python
#jak sortowac po wieku?

sorted(lst1, key=get_age)

#ale jest to toporne
```

```python
#aby tego uniknac uzyj wyrażeń lambda

sorted(lst1, key= lambda nazwa_argumentu: nazwa_argumentu[1])       #dwukropek jak w definicji funkcji
```

```python
# co ciekawe

xD = lambda v: v[1]

xD(('cośtam', 35))
```

# wracamy

```python
@dataclass
class Student:
    name: str
    age: int = field(default_factory=lambda: 20)    #kolejny sposob

    #ale tutaj lepiej po prostu dać: = 20

    email: str = ""     # <- jak jeden argument ma wartośc domuślną, to wszystkie późniejsze też muszą

    classes: list[str] = field(default_factory=lambda: [])
```

```python
bobby = Student(name = 'bobby')
alicja = Student(name='alicja')

bobby, alicja
```

```python
bobby.classes = ['matma']
alicja.classes = ['nie matma']

bobby, alicja
```

# walidacja

```python
@dataclass
class Student:
    name: str
    age: int
    email: str
    classes: list[str] = field(default_factory=list)

    def __post_init__(self):
        if '@' not in self.email:
            raise ValueError('wpisałeś zły mail IDIOTO')    # raise wywala błąd
```

```python
bobktorystam = Student(name='bob', age=0, email='bob2example.com', classes=[])
```
