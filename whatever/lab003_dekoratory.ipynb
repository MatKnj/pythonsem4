```python
from time import perf_counter
import numpy as np
```

wariant C

```python
class TimerDecorator:
    def __init__(self, func):

        self.func = func
        self.call_counter = 0
        self.call_times = []

    def __call__(self, *args, **kwargs):

        start = perf_counter()
        res = self.func(*args, **kwargs)
        end = perf_counter()

        self.call_counter += 1
        self.call_times.append(end - start)

        return res

    def print_stats(self):

        print(f'Nazwa funkcji: {self.func.__name__}')
        print(f'Liczba wywołań: {self.call_counter}')
        print(f'Najkrótszy czas wykonania: {np.min(self.call_times)} s')
        print(f'Najdłuższy czas wykonania: {np.max(self.call_times)} s')
        print(f'Średni czas wykonania: {np.mean(self.call_times)} s')

    def give_stats(self):
        return {'nazwa funkcji' : self.func.__name__, 'liczba wykonań' : self.call_counter,'najkrótszy czas wykonania' : np.min(self.call_times),'najdłuższy czas wykonania' : np.max(self.call_times), 'średni czas wykonania' : np.mean(self.call_times)}



##################################
    

@TimerDecorator
def test_func(x, y, z):
    return x*y*z


@TimerDecorator
def test2_func(x, y):
    return x**y


print('FUNKCJA 1:')
print(test_func(3, 4, 5))
print(test_func(3, 4, 5))
print(test_func(3, 4, 5))


print('FUNKCJA 2:')
print(test2_func(3, 4))
print(test2_func(3, 4))


print('FUNKCJA 1, STATYSTYKI:')
test_func.print_stats()
print(test_func.give_stats())

print('FUNKCJA 2, STATYSTYKI:')
test2_func.print_stats()
print(test2_func.give_stats())
```

```python
class TimerDecorator:
    def __init__(self, show_call_time = False):

        self.show_call_time = show_call_time
        self.call_counter = 0
        self.call_times = []
        self.func = None

    def __call__(self, func):

        self.func = func

        def wrapper(*args, **kwargs):
            start = perf_counter()

            res = self.func(*args, **kwargs)

            end = perf_counter()

            self.call_counter += 1
            self.call_times.append(end - start)

            if self.show_call_time:
                print(f'Czas wywołania: {end - start} s')

            return res
        wrapper.print_stats = self.print_stats
        wrapper.give_stats = self.give_stats
        return wrapper


    def print_stats(self):

        print(f'Nazwa funkcji: {self.func.__name__}')
        print(f'Liczba wywołań: {self.call_counter}')
        print(f'Najkrótszy czas wykonania: {np.min(self.call_times)} s')
        print(f'Najdłuższy czas wykonania: {np.max(self.call_times)} s')
        print(f'Średni czas wykonania: {np.mean(self.call_times)} s')

    def give_stats(self):
        return {'nazwa funkcji' : self.func.__name__, 'liczba wykonań' : self.call_counter,'najkrótszy czas wykonania' : np.min(self.call_times),'najdłuższy czas wykonania' : np.max(self.call_times), 'średni czas wykonania' : np.mean(self.call_times)}


    
#################################################



@TimerDecorator(show_call_time = True)
def test_func(x, y, z):
    return x*y*z


@TimerDecorator()
def test2_func(x, y):
    return x**y


print('FUNKCJA 1:')
print(test_func(3, 4, 5))
print(test_func(3, 4, 5))
print(test_func(3, 4, 5))


print('FUNKCJA 2:')
print(test2_func(3, 4))
print(test2_func(3, 4))




print('FUNKCJA 1, STATYSTYKI:')
test_func.print_stats()
print(test_func.give_stats())

print('FUNKCJA 2, STATYSTYKI:')
test2_func.print_stats()
print(test2_func.give_stats())
```

```python
class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius

    @property
    def celsius(self):
        return self._celsius

    @property
    def kelvin(self):
        k = self._celsius + 273.15
        return k

    @property
    def fahrenheit(self):
        f = 32 + (9 * self._celsius/ 5)
        return f
    
    @celsius.setter
    def celsius(self, c):
        self._celsius = c

        if self.kelvin < 0:
            self.kelvin = 0

    @fahrenheit.setter
    def fahrenheit(self, f):
        c = (f - 32) * (5 / 9)
        self._celsius = c

        if self.kelvin < 0:
            self.kelvin = 0


    @kelvin.setter
    def kelvin(self, k):
        K = k

        if k < 0:
            K = 0

        c = K - 273.15
        self._celsius = c


    def __str__(self):

        return f'Temperatura: {self._celsius} *C'
    

    @property
    def is_freezing(self):
        if self._celsius <= 0:
            return True
        
        else:
            return False
        
    @property
    def is_boiling(self):
        if self._celsius >= 100:
            return True
        
        else:
            return False
        
    @property
    def state(self):
        if self.is_freezing:
            return 'solid'
        
        elif self.is_boiling:
            return 'gas'
        
        else:
            return 'liquid'



########################
t = Temperature(25)

print(t.celsius)      # 25
print(t.fahrenheit)   # 77.0
print(t.kelvin)       #298.15


t.fahrenheit = 32
print(t.celsius)      # 0.0

t.kelvin = 300
print(t.celsius)      # 26.85

t.kelvin = -300      
print(t.kelvin)      # 0

t.celsius = -300      
print(t.celsius, t.kelvin)      # -275.15 = 0

t.fahrenheit = -3000      
print(t.fahrenheit, t.kelvin)      # -459.67 = 0


print(t)

print(t.is_boiling, t.is_freezing, t.state)

t.celsius = 25
print(t.is_boiling, t.is_freezing, t.state)
```

```python
def los_gen(seed, a, c, m, N):

    x = seed

    for i in range(N):
        x2 = (a*x + c) % m
        x = x2
        yield x
```

```python
for i in los_gen(1, 5, 3, 16, 10):
    print(i)    
```
