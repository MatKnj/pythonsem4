```python
class Counter:
    def __init__(self, val):
        self.val = val

    def plus(self):
        self.val += 1

    def minus(self):
        self.val -= 1

    def zero(self):
        self.val = 0

    def read(self):
        print(self.val)



```

```python
licznik = Counter(0)

licznik.read()

for i in range(10):
    licznik.plus()
    licznik.read()


for i in range(4):
    licznik.minus()
    licznik.read()


licznik.zero()
licznik.read()
```

```python
import numpy as np
```

```python
from dataclasses import dataclass  
```

```python
@dataclass(slots=True)
class Measurment:
    time: int
    value: float
    unit: str
```

```python
m1 = Measurment(0, 13, 'm')
m2 = Measurment(1, 17, 'm')
m3 = Measurment(2, 16, 'm')
m4 = Measurment(3, 15, 'm')

meas = [m1, m2, m3, m4]
max = meas[0].value

for i in meas:
    print(f'czas pomiaru: {i.time}, wartość: {i.value} {i.unit}')

    if i.value > max:
        max = i.value


print(f'Największy pomiar: {max} m')


```

```python
class Vector2D:
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y
    
    def __str__(self):
        return f'[{self.x}, {self.y}]'
    
    def __add__(self, vec2):
        return Vector2D(self.x + vec2.x, self.y + vec2.y)
    
    def len(self):
        return np.sqrt(self.x**2 + self.y**2)
    
    def __mul__(self, val):
        return Vector2D(self.x * val, self.y * val)
    
```

```python
wek1 = Vector2D(1, 2)

print(wek1)

wek2 = Vector2D(1, 0)

wek3 = wek1 + wek2

print(wek3)

wek4 = wek3 * wek3.len()
print(wek4)
```

```python
@dataclass(slots = True)
class MeasurmentSeries:
    lst: list[float]

    def __getitem__(self, key):
        return self.lst[key]
    
    def __setitem__(self, key, value):
        self.lst[key] = value

    def __len__(self):
        return len(self.lst)
    
    def __iter__(self):
        for item in self.lst:
            yield item

    def mean(self):
        return np.mean(self.lst)
```

```python
meser1 = MeasurmentSeries([10, 12, 14, 13, 11, 10])

print(meser1[0])

meser1[0] = 11

print(meser1[0])

print(len(meser1))

for i in meser1:
    print(i)

print(meser1.mean())
```

```python
@dataclass(slots = True)
class Polynomial:
    wsp: list[float]

    def __str__(self):
        wiel = ''

        for i in range(len(self.wsp)):
            if i == 0 and self.wsp[i] != 0:
                wiel += str(self.wsp[0])

            elif self.wsp[i] != 0 and self.wsp[i] > 0:
                wiel += ' + ' + str(self.wsp[i]) + 'x**' + str(i)

            elif self.wsp[i] != 0 and self.wsp[i] < 0:
                wiel += ' - ' + str(-self.wsp[i]) + 'x**' + str(i)

        return wiel
    
    def stopien(self):
        return len(self.wsp) - 1
    

    def __call__(self, x):
        y = 0

        for i in range(len(self.wsp)):
            y += self.wsp[i] * (x**i)

        return y
```

```python
wiel1 = Polynomial([1, 0, 0, 2, -4])

print(wiel1)

print(wiel1.stopien())

print(wiel1(1))
```

```python
@dataclass(slots=True)
class Account:
    name: str
    points: int

    def add_points(self, val):
        self.points += val

    def remove_points(self, val):
        self.points -= val
    
    def info(self):
        print(f'nazwa konta: {self.name}, liczba punktów: {self.points}')

```

```python
@dataclass(slots=True)
class AccountPremium(Account):
    premia: float

    def __init__(self, name, points, premia):
        super().__init__(name, points)
        self.premia = premia

    def add_points(self, val):
        return super().add_points(val * self.premia)
```

```python
premacc1 = AccountPremium('konto1', 100, 1.5)

premacc1.info()

premacc1.add_points(10)

premacc1.info()

premacc1.remove_points(10)

premacc1.info()
```

```python
@dataclass(slots=True)
class Animal:
    name: str
    x: int
    y: int

    def move(self):
        pass

    def sound(self):
        pass

    def info(self):
        print(f'nazwa: {self.name}, pozycja: x = {self.x}, y = {self.y}')
```

```python
class Dog(Animal):

    def __init__(self, name, x, y):
        super().__init__(name, x, y)

    def move(self, x, y):
        if abs(self.x - x) < 2:
            self.x = x
        
        if abs(self.y - y) < 2:
            self.y = y
    
    def sound(self):
        print(f'{self.name} zaszczekał')
```

```python
class Bird(Animal):

    def __init__(self, name, x, y):
        super().__init__(name, x, y)

    def move(self, x, y):
        self.x = x
        self.y = y
    
    def sound(self):
        print(f'{self.name} zaćwierkał')
```

```python
pies = Dog('Reksio', 0, 0)
ptak = Bird('Wróbel', 0, 0)

zoo = np.array([pies, ptak])


for i in zoo:
    i.sound()

for i in zoo:
    i.move(1, 1)
    i.info()

for i in zoo:
    i.move(5, 5)
    i.info()

for i in zoo:
    i.sound()
```
