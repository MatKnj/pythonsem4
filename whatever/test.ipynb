```python
class ElectronicDevice:
  def __init__(self, imie=''):
    print('Creating Elec Dev')
    self.name = imie

  def restart(self):
    print(f'Restart {self.name}')

  __restart = restart

  def update_firmware(self):
    print(f'update firmware for {self.name}')
    self.__restart()


class Computer(ElectronicDevice):
  def __init__(self, name, ram):
    print('creating a komputer')
    super().__init__(name)
    self.ram = ram

  def restart(self):
    print(f'saving files on {self.name}')
    super().restart()


class Camera(ElectronicDevice):
  def __init__(self, name, width, height):
    print("Creating Camera")
    super().__init__(name)
    self.width = width
    self.height = height

  def restart(self):
    print(f"Camera restart, backing up photos on {self.name}")
    super().restart()
```
